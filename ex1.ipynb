```python
t = Table.read("hst_results_nd.fits")
ra = t["RA"]
dec = t["DEC"]
av = t["Av_p50"]
age = t["logA_p50"]
mass=t["M_ini_p50"]
temp = t["logT_p50"]
f475w = t["F475_VEGA"]
f418w = t["F814_VEGA"]
#t[0].colnames
```

```python
young_stars = age (age < 8.0)
old_strs = age(age > 9.5)
#f475w_ys = f475[young]
#print(old_stars[0:100])
plt.figure()
plt.plot(f475w  - f814w, f475w, ',', color = 'gray'. ls = ''label = "all stars")
plt.ylim(26.5, 17.3) 
plt.plot (young_stars["F475_VEGA"] - young_stars ["F814_VEGA"], young_stars["F475_VEGA"], 'b.', ls = '', label='log(age) < 8.5 dex')
plt.plot (old_stars["F475_VEGA"] - old_stars ["F814_VEGA"], old_stars["F475_VEGA"], 'r.', ls = '', label='log(age) > 10 dex')
plt.legend()
plt.savefig("cmd_by_age.pdf")
```
